```python
import pandas as pd

def aggregate_features_account_time(df, acct_col):
    # æ ¹æ“šå¸³æˆ¶æ±ºå®šå°æ‡‰çš„ type æ¬„ä½
    type_col = 'from_acct_type' if acct_col == 'from_acct' else 'to_acct_type'

    # ğŸ”¹ æ ¹æ“šæ™‚é–“å€é–“å»ºç«‹çµ±è¨ˆæ¬„
    df['txn_0_6'] = ((df['txn_hours'] >= 0) & (df['txn_hours'] < 6)).astype(int)
    df['txn_6_12'] = ((df['txn_hours'] >= 6) & (df['txn_hours'] < 12)).astype(int)
    df['txn_12_18'] = ((df['txn_hours'] >= 12) & (df['txn_hours'] < 18)).astype(int)
    df['txn_18_24'] = ((df['txn_hours'] >= 18) & (df['txn_hours'] <= 24)).astype(int)

    # ğŸ”¹ å°å¸³æˆ¶æ•´é«”èšåˆ
    agg = df.groupby(acct_col).agg(
        txn_amt_max=('txn_amt_twd', 'max'),
        txn_amt_min=('txn_amt_twd', 'min'),
        txn_amt_mean=('txn_amt_twd', 'mean'),
        txn_count=('txn_amt_twd', 'count'),
        txn_0_6=('txn_0_6', 'sum'),
        txn_6_12=('txn_6_12', 'sum'),
        txn_12_18=('txn_12_18', 'sum'),
        txn_18_24=('txn_18_24', 'sum'),
        type_1=(type_col, lambda x: (x == 1).sum()),
        type_2=(type_col, lambda x: (x == 2).sum()),
        ch_type_1=('channel_type', lambda x: (x == '1').sum() if x.dtype == object else (x == 1).sum()),
        ch_type_0=('channel_type', lambda x: (x == '0').sum() if x.dtype == object else (x == 0).sum())
    ).reset_index()

    return agg

# ä½¿ç”¨æ–¹å¼
df_from_features = aggregate_features_account_time(df_from, 'from_acct')
df_to_features   = aggregate_features_account_time(df_to, 'to_acct')

# æª¢è¦–çµæœ
print(df_from_features.head())
print(df_to_features.head())

```

```python
# å…ˆæŠŠ from_acct_features èˆ‡ to_acct_features æ¬„ä½åç¨±å€åˆ†é–‹
df_from_features_renamed = df_from_features.rename(
    columns=lambda x: 'from_' + x if x != 'from_acct' else 'acct'
)

df_to_features_renamed = df_to_features.rename(
    columns=lambda x: 'to_' + x if x != 'to_acct' else 'acct'
)

# ä»¥å¸³è™Ÿç‚º key é€²è¡Œå¤–é€£æ¥ï¼ˆouter joinï¼‰
df_account_features = pd.merge(
    df_from_features_renamed,
    df_to_features_renamed,
    on='acct',
    how='outer'
)

# æª¢è¦–çµæœ
print(df_account_features.head())
```

```python
import pandas as pd

# å°‡ NaN ç•¶ä½œ 0
df_from_features = df_from_features.fillna(0)
df_to_features   = df_to_features.fillna(0)

# é‡æ–°å‘½åæ¬„ä½ä»¥ä¾¿å€åˆ†
df_from_features_renamed = df_from_features.rename(
    columns=lambda x: 'from_' + x if x != 'from_acct' else 'acct'
)
df_to_features_renamed = df_to_features.rename(
    columns=lambda x: 'to_' + x if x != 'to_acct' else 'acct'
)

# å¤–é€£æ¥åˆä½µ
df_merged = pd.merge(df_from_features_renamed, df_to_features_renamed, on='acct', how='outer').fillna(0)

# å»ºç«‹æœ€çµ‚æ•´åˆ DataFrame
df_final = pd.DataFrame()
df_final['acct'] = df_merged['acct']

# äº¤æ˜“é‡‘é¡ max/min
df_final['txn_amt_max'] = df_merged[['from_txn_amt_max', 'to_txn_amt_max']].max(axis=1)
df_final['txn_amt_min'] = df_merged[['from_txn_amt_min', 'to_txn_amt_min']].min(axis=1)

# å¹³å‡äº¤æ˜“é‡‘é¡åŠ æ¬Šå¹³å‡
df_final['txn_amt_mean'] = (
    df_merged['from_txn_amt_mean'] * df_merged['from_txn_count'] +
    df_merged['to_txn_amt_mean'] * df_merged['to_txn_count']
) / (df_merged['from_txn_count'] + df_merged['to_txn_count']).replace(0, 1)  # é¿å…é™¤ä»¥ 0

# äº¤æ˜“ç­†æ•¸ç›¸åŠ 
df_final['txn_count'] = df_merged['from_txn_count'] + df_merged['to_txn_count']

# æ™‚æ®µäº¤æ˜“æ¬¡æ•¸ç›¸åŠ 
time_cols = ['txn_0_6', 'txn_6_12', 'txn_12_18', 'txn_18_24']
for col in time_cols:
    df_final[col] = df_merged['from_' + col] + df_merged['to_' + col]

# é¡å‹å’Œé€šè·¯æ¬„ä½ç›¸åŠ 
add_cols = ['type_1', 'type_2', 'ch_type_1', 'ch_type_0']
for col in add_cols:
    df_final[col] = df_merged['from_' + col] + df_merged['to_' + col]

# æª¢è¦–çµæœ
print(df_final.head())


```

```python
import pandas as pd

# è®€å–é æ¸¬å¸³è™Ÿ
predict_accts = predictdf_csv['acct']

# å¾ df_final ç§»é™¤é€™äº›å¸³è™Ÿåšè¨“ç·´é›†
df_train = df_final[~df_final['acct'].isin(predict_accts)].copy()

# å°‡è¢«ç§»é™¤çš„å¸³è™Ÿå­˜åˆ° predictss
predictss = df_final[df_final['acct'].isin(predict_accts)].copy()

# æª¢æŸ¥çµæœ
print("è¨“ç·´é›†å¤§å°:", df_train.shape)
print(df_train.head())
print("é æ¸¬é›†å¤§å°:", predictss.shape)
print(predictss.head())
```

```python
# è®€å–è­¦ç¤ºå¸³è™Ÿ
alert_accts = alertdf_csv['acct']

# æ¨™è¨» label
df_train['label'] = df_train['acct'].apply(lambda x: 1 if x in set(alert_accts) else 0)
predictss['label'] = predictss['acct'].apply(lambda x: 1 if x in set(alert_accts) else 0)

# æª¢æŸ¥
print(df_train['label'].value_counts())
print(predictss['label'].value_counts())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# å‡è¨­ df_train å·²ç¶“æœ‰ features + label
X_train = df_train.drop(columns=['label']==0)
X_val = df_train['label']
y_train = df_train.drop(columns=['label']==1)
y_val = df_train['label']

# åˆ†è¨“ç·´/é©—è­‰é›†
X_train, X_val, y_train, y_val 

# å»ºç«‹æ¨¡å‹
models = {
    'logistic': LogisticRegression(max_iter=1000),
    'decision_tree': DecisionTreeClassifier(random_state=42),
    'random_forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'xgboost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# è¨“ç·´æ¨¡å‹ä¸¦é©—è­‰
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_val)
    print(f"===== {name} =====")
    print("Accuracy:", accuracy_score(y_val, preds))
    print(classification_report(y_val, preds))

# é›†æˆæ¨¡å‹ (Voting)
voting_clf = VotingClassifier(
    estimators=[(name, mdl) for name, mdl in models.items()],
    voting='soft'  # soft voting ç”¨é æ¸¬æ©Ÿç‡å¹³å‡
)
voting_clf.fit(X_train, y_train)
preds_vote = voting_clf.predict(X_val)
print("===== Voting Ensemble =====")
print("Accuracy:", accuracy_score(y_val, preds_vote))
print(classification_report(y_val, preds_vote))

# å° df_predict åšé æ¸¬
X_predict = predictss.drop(columns=['label'])
predictdf_csv['label'] = voting_clf.predict(X_predict)
```
